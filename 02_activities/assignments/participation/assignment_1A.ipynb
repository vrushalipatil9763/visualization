```python
# Import standard libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#define X and Y axis
np.random.seed(613)
X = np.arange(50)
y = np.random.randint(0, 100, 50)

#scatter plot 
fig, ax = plt.subplots(figsize = (5,3))       #Figsize = size of graph in inches
ax.scatter(X,y)
```

```python
#bar plot 
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(X,y)
```

```python
#horizontal bar plot 
fig, ax = plt.subplots(figsize = (5,3))
ax.barh(X,y)
```

```python
#Histogram plot 
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
#use of different font objects
font1 = {'family' : 'sans-serif', 'color' : 'blue', 'size':20}
font2 = {'family' : 'monospace', 'color' : 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(X,y)
#Set plot titles
ax.set_title("Total growth over time", fontdict= font1, loc = 'left', weight = 'bold')
ax.set_ylabel("Total growth", fontdict= font2)
ax.set_xlabel("Years since start", fontdict= font2)
```

```python
#Set color of plot data
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(X, y, marker = '*', color = 'indigo')
```

```python
#use of color codes cat codes, set edge color, line color, line sryle
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(X, y, marker = '*', 
        color = "#8E2397", 
        linestyle = '--', 
        linewidth = 2,
        markersize = 12,
        markeredgecolor = '#FA9388',
        markerfacecolor = '#000000')

#draw vertical and horizontal grid lines
ax.grid(axis = 'y',  color = "#BACA2D", linewidth = 2, linestyle = '-')
ax.grid(axis = 'x',  color = "#CA2D83", linewidth = 2, linestyle = '-')
```

```python
#create a bubble plot
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')
data['continent'] = pd.Categorical(data['continent'])
data.head()
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

# Subset of the data for year 1952
data1952 = data[data.year == 1952]

# Scatterplot
ax.scatter(
    x = data1952['lifeExp'], 
    y = data1952['gdpPercap'], 
    s = data1952['pop']/50000, 
    c = data1952['continent'].cat.codes, 
    cmap = "Accent", 
    alpha = 0.6, 
    edgecolors = "white", 
    linewidth = 2
)
 
# Add titles (main and on axis)
plt.yscale('log')
ax.set_xlabel("Life Expectancy")
ax.set_ylabel("GDP per Capita")
ax.set_title("Year 1952")
ax.set_ylim(100,50000)
ax.set_xlim(30, 75)

plt.show()
```
