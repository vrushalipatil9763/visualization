```python
#Use of different transforms
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL 
import requests

np.random.seed(613)
X = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
##plot both y variables on the same axis
plt.style.use('default')
fig, ax = plt.subplots(figsize = (5,3))       #Figsize = size of graph in inches
ax.plot(X,y1, label = "Person 1", color = "blue")
ax.plot(X,y2, label = "Person 2", color = "orange")
ax.legend(loc = 'lower right',
          frameon= True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
```

```python
#add legend outside of graph
fig, ax = plt.subplots(figsize = (5,3))  
ax.plot(X,y1, label = "Person 1", color = "blue")
ax.plot(X,y2, label = "Person 2", color = "orange")
ax.legend(loc = 'upper left',
          bbox_to_anchor = (1,1))

```

```python
#Text location

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(X, y1, label = "Person 1")
ax.scatter(X, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.text(10,95, "This value is important",
        ha = 'center',
        color = 'red',
        size =20
        )
```

```python
#Use of different transforms
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform = ax.transData)
#ax.text(0.5, 0.1, ". Data: (.5, .1)", transform = ax.transData)

ax.text(0.5, 0.1, ".Axes:(0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2, ".Figure:(0.2, 0.2)", transform = fig.transFigure)
```

```python
#Use of arrow
fir, ax = plt.subplots(figsize  = (5,3))
ax.scatter(X, y1, label = "Person 1")
ax.scatter(X, y2, label = "Person 2")
ax.legend(loc = "upper left",
          bbox_to_anchor = (1,1))

ax.annotate("This is important", xy = (10,95), xytext = (20,94), 
            arrowprops = dict(facecolor = 'black',
                              arrowstyle = "fancy"))
```

```python
#X,Y axis defintion
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110,180,240,99,220])
```

```python
#adjust all the labels within plot area

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows= 1,
                               figsize = (10,3))
ax1.scatter(x1,y1, marker = '*')
ax2.bar(x2,y2, color ="green")

ax1.set_title('Scatterplot', fontdict=font1)
ax1.set_ylabel("measure", fontdict=font1)
ax1.set_xlabel("Time", fontdict=font1)

ax2.set_title('Prirate Crew Strength', fontdict=font1)
ax2.set_ylabel("strength", fontdict=font1)
ax2.set_xlabel("Charater", fontdict=font1)
```

```python
#use of mosaic style
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize = (7,4),
                                    layout = "constrained")

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel("A big Label", fontsize = 18)
someaxes['ax2'].set_xlabel("Another Label", fontsize = 18)
someaxes['ax3'].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize = 18)
```
